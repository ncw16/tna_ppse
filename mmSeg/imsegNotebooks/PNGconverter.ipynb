## Convert prize paper images into PNG files

```python
# Import libraries
import os
import cv2
```

```python
directory = "../SampleSets"

savedir = "../SampleSetsPNG"

try:
    os.mkdir(savedir)
except OSError:
    print ("Creation of the directory %s failed" % savedir)
else:
    print ("Successfully created the directory %s " % savedir)

print("Searching for files")
for dirName, subdirList, fileList in os.walk(directory):
    for fname in fileList:
        print('\t%s' % fname)
        
        #read the image
        img = cv2.imread(dirName + '/' + fname, cv2.IMREAD_UNCHANGED)
        cv2.imwrite('../SampleSetsPNG/' + fname[0:-4] + '.png', img)
        print('Sucessful save')
```

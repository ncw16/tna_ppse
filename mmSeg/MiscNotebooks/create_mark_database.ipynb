```python
# Import libraries
import os
import pandas as pd
import json
import csv
from PIL import Image
```

## Setup folders, paths, read CSV

```python
# Load the ground truth csv file
via_export = pd.read_csv("../MarkupFromVIA/marks_markup/via_export_csv.csv")

# Create the directory for the mark test database
path = "../gtmark_test_db/"

try:
    os.mkdir(path)
except OSError:
    print ("Creation of the directory %s failed" % path)
else:
    print ("Successfully created the directory %s " % path)

sample_dir = "../SampleSets/"
```

## Extract the mark label

```python
# Cycle through each instance of a mark
mark_labels = []
for x in range(via_export.shape[0]):
    
    # Extract the file name and mark properties
    data = via_export.iloc[x]
    print("\nMark information (From csv)")
    print(data)
    
    # mark_dimensions = [left,right,upper,lower coordinates]
    mark_dims = []
    temp_shape_data = data.region_shape_attributes[1:-1].split(',')
    
    # Check that the image actually has a mark (5 consists of the name,x,y,width,height 
    #                                           where x,y are the bottom left coordinates)
    if len(temp_shape_data) < 5:
        mark_labels.append("")
        continue
    else:
        temp_dims = []
        for y in range(1,5):
            value = temp_shape_data[y].split(':')
            
            temp_dims.append(int(value[1]))
        
        # Fill the mark_dims array [left, right, upper, lower]
        mark_dims.append(temp_dims[0])
        mark_dims.append(temp_dims[0]+temp_dims[2])
        mark_dims.append(temp_dims[1])
        mark_dims.append(temp_dims[1]+temp_dims[3])
        
        print("\nMark dims:")
        print(mark_dims)
        
        # Open the original image
        im = Image.open(sample_dir + data.filename)
        
        # Crop the image
        #imCrop = im.crop((mark_dims[0], mark_dims[2], mark_dims[1], mark_dims[3]))
        
        img_dir = path + data.filename[0:-4] + "_" + str(x+2) + ".jpg"
        
        # Save the cropped image
        #imCrop.save(img_dir, "JPEG")

        # Get the mark label
        mark_info = data.region_attributes[1:-1].split(',')
        temp_data = mark_info[0].split(':')
        m_label = temp_data[1][1:-1]
        
        print("Label:")
        print(m_label)
        
        mark_labels.append(m_label)
```

## Create a new CSV for the groundtruth mark database, used to test DocRetrieval component

```python
# Create new csv
print(mark_labels)
print("\nThe number of mark labels:")
print(len(mark_labels))

print("\nCreating new CSV data frame")
gtruth_db = via_export.assign(mark_label = mark_labels)

print("\nSaving")

gtruth_db.to_csv(path + "gtruthDB.csv", index=True)

print("\nCompleted\n")

print(gtruth_db)
```

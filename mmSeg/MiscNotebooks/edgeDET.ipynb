## Edge Detection practice

```python
import os
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Location of sample sets
rootDir = '../ResizedSS4/'

# Create the directory for the ouptut
path = "../Outputs/edge_images/"

try:
    os.mkdir(path)
except OSError:
    print ("Creation of the directory %s failed" % path)
else:
    print ("Successfully created the directory %s " % path)

    
for dirName, subdirList, fileList in os.walk(rootDir):
    print('Found directory: %s' % dirName)
    for fname in fileList:
        print('\t%s' % fname)

        img = cv2.imread(dirName + '/' + fname)
        
        gry = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        
        edges = cv2.Canny(gry,100,200)

        cv2.imwrite(path + fname, edges)
        print('Sucessful save')
```
